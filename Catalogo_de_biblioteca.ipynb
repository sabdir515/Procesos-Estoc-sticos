```python
import sympy as sp
import networkx as nx
import matplotlib.pyplot as plt
```

3. Representación de matrices

En SymPy las matrices se representan con la función Matrix().

```python
A = sp.Matrix([
    [1, 2],
    [3, 4]
])

B = sp.Matrix([
    [0, 1],
    [1, 0]
])

A, B
```

4. Obtener una entrada, fila o columna

Las matrices en Python empiezan a contarse desde 0.

Obtener una entrada específica

Esto obtiene el elemento en fila 0, columna 1.

```python
A[0,1]
```

Obtener una fila

```python
A.row(1)
```

Obtener una columna

```python
A.col(0)
```

5. Multiplicación de matrices

La multiplicación se realiza con el operador *.

```python
C = A * B
C
```

6. Elevar una matriz a una potencia

Para elevar una matriz se usa **.

```python
A2 = A**2
A2
```

También se puede calcular:

```python
A**3
```

7. Multiplicación de un vector por una matriz

Un vector también se puede representar como matriz columna.

```python
v = sp.Matrix([
    [1],
    [2]
])

resultado = A * v
resultado
```

8. Valores y vectores propios

Los valores propios (eigenvalues) y vectores propios se obtienen con eigenvects().

Celda de código

```python
A.eigenvects()
```

El resultado contiene:

* valor propio

* multiplicidad

* vector(es) propio(s)

Si solo se quieren los valores propios:

```python
A.eigenvals()
```

9. Diagramas de transición con NetworkX

Un diagrama de transición se usa en procesos estocásticos o cadenas de Markov, donde los nodos representan estados y las flechas representan probabilidades de transición.

Con NetworkX se puede construir fácilmente.

```python
G = nx.DiGraph()
```

Agregar estados

```python
G.add_nodes_from(["A","B"])
```

Agregar transiciones

```python
G.add_edge("A","A", weight=0.7)
G.add_edge("A","B", weight=0.3)
G.add_edge("B","A", weight=0.4)
G.add_edge("B","B", weight=0.6)
```

10. Dibujar el diagrama de transición
Celda de código

```python
pos = nx.spring_layout(G)

nx.draw(G, pos, with_labels=True, node_color="lightblue", node_size=2000)

labels = nx.get_edge_attributes(G,'weight')
nx.draw_networkx_edge_labels(G, pos, edge_labels=labels)

plt.show()
```
