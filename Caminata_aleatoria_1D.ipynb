# Introducción

En esta práctica se implementa una caminata aleatoria unidimensional (1D).

En cada paso, la partícula puede moverse:

- Una unidad a la derecha con probabilidad $p$.
- Una unidad a la izquierda con probabilidad $q$.

Cumpliendo la condición:

$$
p + q = 1
$$

Esto garantiza que en cada paso la partícula necesariamente se moverá en alguna de las dos direcciones.

La posición de la partícula después de $n$ pasos se define como:

$$
X_n = \sum_{i=1}^{n} Y_i
$$

donde cada $Y_i$ representa el desplazamiento en el paso $i$ y toma valores:

$$
Y_i =
\begin{cases}
+1 & \text{con probabilidad } p \\
-1 & \text{con probabilidad } 1-p
\end{cases}
$$

El objetivo de esta práctica es:

1. Simular computacionalmente la caminata aleatoria.
2. Visualizar gráficamente la trayectoria de la partícula.
3. Analizar cómo el valor de $p$ influye en el comportamiento del sistema.




```python
import numpy as np
import matplotlib.pyplot as plt
import random as r

def caminata(p=0.5, n=100):
    T = []
    x = 0

    for i in range(n):
        if r.random() < p:
            x += 1
        else:
            x -= 1
        T.append(x)

    return T


def graficar_caminata(p=0.5, n=100):
    trayectoria = caminata(p, n)
    pasos = np.arange(1, n+1)

    plt.figure(figsize=(8,5))
    plt.plot(pasos, trayectoria, marker='o', linestyle='-')
    plt.axhline(0, color='black', linewidth=1)
    plt.title("Trayectoria de una Caminata Aleatoria")
    plt.xlabel("Número de pasos")
    plt.ylabel("Posición")
    plt.grid(True)
    plt.show()


# Llamar la función
graficar_caminata(0.2, 100)
```

# Análisis Matemático de la Caminata Aleatoria 1D

Sea la caminata aleatoria definida por

$$
X_n = \sum_{i=1}^{n} Y_i
$$

donde

$$
Y_i =
\begin{cases}
+1 & \text{con probabilidad } p \\
-1 & \text{con probabilidad } 1-p
\end{cases}
$$

---

## 1. ¿Qué debe pasar para que la partícula regrese al origen?

Para que la partícula regrese al origen en el paso $n$, debe cumplirse:

$$
X_n = 0
$$

Esto implica que el número de pasos hacia la derecha debe ser igual al número de pasos hacia la izquierda.

Si $k$ es el número de pasos hacia la derecha, entonces:

$$
k = \frac{n}{2}
$$

Por lo tanto, $n$ debe ser par.

La probabilidad de regresar al origen es:

$$
P(X_n = 0) = \binom{n}{n/2} p^{n/2} (1-p)^{n/2}
$$
Distrbución Binomial

---

## 2. ¿Cuál es la esperanza matemática?

Primero calculamos la esperanza de un paso:

$$
E[Y_i] = (+1)p + (-1)(1-p)
$$

$$
E[Y_i] = 2p - 1
$$

Como los pasos son independientes:

$$
E[X_n] = n(2p - 1)
$$

Observaciones:

- Si $p = 0.5$, entonces $E[X_n] = 0$.
- Si $p > 0.5$, la caminata tiene tendencia positiva.
- Si $p < 0.5$, la caminata tiene tendencia negativa.
Estas observaciones las podemos corroboara en la gráfica que programamos.

---

## 3. ¿Cuál es la varianza?

Sabemos que:

$$
Var(Y_i) = E[Y_i^2] - (E[Y_i])^2
$$

Como $Y_i^2 = 1$, entonces:

$$
E[Y_i^2] = 1
$$

Por lo tanto:

$$
Var(Y_i) = 1 - (2p - 1)^2
$$

Simplificando:

$$
Var(Y_i) = 4p(1-p)
$$

Dado que los pasos son independientes:

$$
Var(X_n) = 4np(1-p)
$$

---

## 4. ¿Cuál es el promedio de la partícula?

El promedio después de $n$ pasos es:

$$
\frac{X_n}{n}
$$

Por la Ley de los Grandes Números:

$$
\frac{X_n}{n} \to 2p - 1 \quad \text{cuando } n \to \infty
$$

---

## 5. ¿Qué ocurre cuando $n$ es muy grande?

Para $n$ grande, por el Teorema Central del Límite:

$$
X_n \approx \mathcal{N}\big(n(2p-1), \, 4np(1-p)\big)
$$

Es decir, la distribución se aproxima a una normal.

---

## 6. ¿Cómo influye la diferencia entre $p$ y 0.5?

Mientras mayor sea:

$$
|p - 0.5|
$$

mayor será la tendencia o desplazamiento promedio de la caminata.

Si $p = 0.5$, la caminata es simétrica.
Si $p \neq 0.5$, existe un sesgo o "drift".

---

## 7. ¿Cómo crece la dispersión?

La desviación estándar es:

$$
\sigma = 2\sqrt{np(1-p)}
$$

Esto muestra que la dispersión crece proporcionalmente a $\sqrt{n}$.
